# Data Cleaning and Preprocessing

## Project Objective

The objective of this project is to clean the raw e-commerce dataset by identifying and handling missing values, checking for duplicate records, and verifying data formats. Data cleaning ensures that the dataset is accurate, consistent, and ready for further analysis.

```python
import pandas as pd
```

```python
df = pd.read_excel("Dataset for Data Analytics.xlsx")


df.head()
```

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()

print("Number of Duplicate Rows:", duplicates)
```

```python
df.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:")
print(df.shape)
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

print(df["Date"].dtype)
```

```python
df.describe()
```

```python
# Display cleaned dataset

df.head()
```

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)

print("Cleaned dataset saved successfully!")
```

Conclusion

1. Checked the dataset for missing values.
2. Found 309 missing values in the CouponCode column.
3. Replaced missing CouponCode values with "No Coupon".
4. Checked for duplicate records and found none.
5. Verified data types and converted the Date column to datetime format.
6. Saved the cleaned dataset for further analysis.
